```python
import requests
import zipfile
import io
import pandas as pd

# -------------------------------------
# CONFIG
# -------------------------------------
API_TOKEN = "xapt4GarlMnBKLM0aR0nypi3wWpHlAoTmfuvIro3"
DATA_CENTER = "ca1"          # e.g., "iad1"
SURVEY_ID = "SV_ekS5XcqnjgJdWNo"
FILE_FORMAT = "csv"                        # Alternatives: "tsv", "spss"
# -------------------------------------

BASE_URL = f"https://ca1.qualtrics.com/API/v3/surveys/SV_ekS5XcqnjgJdWNo/export-responses"

headers = {
    "content-type": "application/json",
    "x-api-token": API_TOKEN
}

# Step 1: Create Export
create_resp = requests.post(
    BASE_URL,
    json={"format": FILE_FORMAT},
    headers=headers
)
create_resp.raise_for_status()

progress_id = create_resp.json()["result"]["progressId"]

# Step 2: Poll Until File is Ready
while True:
    check_resp = requests.get(
        f"{BASE_URL}/{progress_id}",
        headers=headers
    )
    check_resp.raise_for_status()
    status = check_resp.json()["result"]["status"]
    print("Status:", status)

    if status == "complete":
        file_id = check_resp.json()["result"]["fileId"]
        break
    elif status == "failed":
        raise Exception("Qualtrics export failed.")

# Step 3: Download the Exported File
download_resp = requests.get(
    f"{BASE_URL}/{file_id}/file",
    headers=headers,
    stream=True
)
download_resp.raise_for_status()

# Step 4: Extract ZIP file from memory and load into Pandas
zipped = zipfile.ZipFile(io.BytesIO(download_resp.content))
csv_filename = zipped.namelist()[0]         # usually just one file
df = pd.read_csv(zipped.open(csv_filename))

print("Downloaded", csv_filename, "with", len(df), "rows.")
print(df.head())
```
