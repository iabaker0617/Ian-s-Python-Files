```python
# China
# =====================================================
# AUTOMATED QUALTRICS EXPORT → EXCEL → EMAIL VIA OUTLOOK
# =====================================================

import requests
import zipfile
import io
import pandas as pd
import time
import os
from datetime import datetime
import win32com.client as win32

# -----------------------------------------------------
# QUALTRICS API CONFIG
# -----------------------------------------------------
API_TOKEN = "xapt4GarlMnBKLM0aR0nypi3wWpHlAoTmfuvIro3"
DATA_CENTER = "ca1"
SURVEY_ID = "SV_5gwXWwNEk1azYdo"
FILE_FORMAT = "csv"

BASE_URL = f"https://{DATA_CENTER}.qualtrics.com/API/v3/surveys/{SURVEY_ID}/export-responses"

headers = {
    "content-type": "application/json",
    "x-api-token": API_TOKEN
}

payload = {
    "format": FILE_FORMAT,
    "useLabels": True
}

# -----------------------------------------------------
# STEP 1: CREATE EXPORT
# -----------------------------------------------------
create_resp = requests.post(
    BASE_URL,
    json=payload,
    headers=headers
)
create_resp.raise_for_status()

progress_id = create_resp.json()["result"]["progressId"]
print(f"Export started. Progress ID: {progress_id}")

# -----------------------------------------------------
# STEP 2: POLL UNTIL EXPORT IS READY
# -----------------------------------------------------
while True:
    check_resp = requests.get(f"{BASE_URL}/{progress_id}", headers=headers)
    check_resp.raise_for_status()

    status = check_resp.json()["result"]["status"]
    print("Status:", status)

    if status == "complete":
        file_id = check_resp.json()["result"]["fileId"]
        break
    elif status == "failed":
        raise Exception("Qualtrics export failed.")

    time.sleep(5)

# -----------------------------------------------------
# STEP 3: DOWNLOAD & LOAD INTO PANDAS
# -----------------------------------------------------
download_resp = requests.get(
    f"{BASE_URL}/{file_id}/file",
    headers=headers,
    stream=True
)
download_resp.raise_for_status()

zipped = zipfile.ZipFile(io.BytesIO(download_resp.content))
csv_filename = zipped.namelist()[0]
df = pd.read_csv(zipped.open(csv_filename))
# Drop the extra Qualtrics header rows (rows 2 and 3)
df = df.iloc[2:].reset_index(drop=True)


print(f"Downloaded {csv_filename} with {len(df)} rows.")

# -----------------------------------------------------
# STEP 4: SAVE TO EXCEL
# -----------------------------------------------------
today = datetime.today().strftime("%Y-%m-%d")
output_dir = r"C:\Temp"  # Change if desired
os.makedirs(output_dir, exist_ok=True)

excel_path = os.path.join(
    output_dir,
    f"China_Alumni_Survey_Data_{today}.xlsx"
)

df.to_excel(excel_path, index=False)
print(f"Excel file saved to: {excel_path}")

# -----------------------------------------------------
# STEP 5: EMAIL THE EXCEL FILE VIA OUTLOOK
# -----------------------------------------------------
outlook = win32.Dispatch('outlook.application')
mail = outlook.CreateItem(0)

mail.To = "iabaker@bu.edu; kjmarco@bu.edu; jonk@bu.edu"
mail.Subject = "China Alumni Survey Respones"
mail.Body = (
    "Hello,\n\n"
    "Here is the most recent extract of the China Alumni Survey responses. \n\n"
    "Cheers,\n"
    "Ian"
)

mail.Attachments.Add(excel_path)
mail.Send()

print("Email sent successfully via Outlook client with attachment!")
```
