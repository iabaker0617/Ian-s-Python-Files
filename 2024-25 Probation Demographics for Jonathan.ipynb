# Import Spring 2025 list of students on probation and inspect

```python
file_path_2 = r"P:\IE Office_Current\Data Requests\2025\Probation demographics for Jonathan Phipps\ASCFall2024.xlsx"
Spring25Prob = pd.read_excel(file_path_2)
Spring25Prob.rename(columns={'Id Number': 'ACPHSID'}, inplace=True)
Spring25Prob=Spring25Prob[Spring25Prob['Decision'].str.contains('Probation', case=False, na=False)]
FallEnrollment = FallEnrollment[~FallEnrollment['Major1'].str.contains("SPEC", na=False)] #the tilde negates the SPEC, meaning drop SPECs
gui = show(Spring25Prob)
```

# Merge Fall 2024 data and Spring 25 probation data

```python
merged_df=pd.merge(FallEnrollment, Spring25Prob, on='ACPHSID', how='left', indicator=True)
merged_df = merged_df[['_merge', 'CollegeLevel', 'Ethniccode', 'Sex', 'Major1', 'Decision']]
merged_df.rename(columns={'Ethniccode': 'Ethnicity'}, inplace=True)
merged_df.rename(columns={'Major1': 'Major 1'}, inplace=True)
merged_df.rename(columns={'Sex': 'Gender'}, inplace=True)
merged_df.rename(columns={'CollegeLevel': 'College Level'}, inplace=True)
merged_df['DecisionRecode'] = merged_df['Decision'].str.contains('Probation', case=False, na=False).astype(int)
merged_df['URM'] = merged_df['Ethnicity'].str.contains('WHITE', case=False, na=False).apply(lambda x: 0 if x else 1)
merged_df['IntlStudent'] = np.where(
    merged_df['Ethnicity'].str.contains('NON RES', case=False, na=False),
    1,
    0
)
gui = show(merged_df)
```

# Chi Square for sex (M vs F)

```python
filtered_df = merged_df[(merged_df['DecisionRecode'] == 1) & (merged_df['Gender'].isin(['F', 'M']))]
sex_tabulation=filtered_df['Gender'].value_counts()
print(sex_tabulation)
```

```python
filtered_df = merged_df[(merged_df['DecisionRecode'] == 1)]
ethnic_tab=filtered_df['Ethnicity'].value_counts()
print(ethnic_tab)
```

```python
from scipy.stats import chi2_contingency
filtered_df = merged_df[merged_df['Gender'].isin(['F', 'M'])]
contingency_table = pd.crosstab(filtered_df['Gender'], filtered_df['DecisionRecode'])
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
N=filtered_df.shape[0]
print(f"Number of observations (N): {N}")
print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of freedom: {dof}")
print(f"Expected frequencies table:\n{expected}")
```

```python
from scipy.stats import fisher_exact
# Filter the data
filtered_df = merged_df[merged_df['Gender'].isin(['F', 'M'])]
# Create contingency table
contingency_table = pd.crosstab(filtered_df['Gender'], filtered_df['DecisionRecode'])
# Perform Fisher's Exact Test
odds_ratio, p_value = fisher_exact(contingency_table)
# Get the number of observations
N = filtered_df.shape[0]  # Number of rows in the filtered dataframe
# Print results
print(f"Number of observations (N): {N}")
print(f"Odds ratio: {odds_ratio}")
print(f"P-value: {p_value}")
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
filtered_df = merged_df[merged_df['Gender'].isin(['F', 'M'])]
# Assuming 'filtered_df' is your DataFrame, and 'DecisionRecode', 'URM', 'Sex' are columns

# Use the formula interface for the regression, specifying categorical variables with C()
formula = 'DecisionRecode ~ C(URM) + C(Gender)'

# Fit the regression model with robust standard errors (HC3)
model = smf.ols(formula, data=filtered_df)
results = model.fit(cov_type='HC3')

# Display the results
print(results.summary())
```
